# Feature 5 – Monthly Trend Analysis

```python
df["Month"] = df["Date"].dt.month_name()
```

```python
debit_df = df[df["Type"] == "debit"]
```

```python
df[["Date", "Month"]].head()
```

```python
monthly_spending = debit_df.pivot_table(
    values="Amount",
    index="Category",
    columns="Month",
    aggfunc="sum",
    fill_value=0
)

monthly_spending
```

```python
month_order = [
    "January",
    "February",
    "March",
    "April",
    "May",
    "June"
]

monthly_spending = monthly_spending.reindex(columns=month_order, fill_value=0)

monthly_spending
```

```python
monthly_spending["Growth (%)"] = (
    (monthly_spending["June"] - monthly_spending["January"])
    / monthly_spending["January"].replace(0, np.nan)
) * 100
```

```python
highest_growth = monthly_spending["Growth (%)"].idxmax()
growth_value = monthly_spending.loc[highest_growth, "Growth (%)"]

print("Highest Growth Category:", highest_growth)
print(f"Growth: {growth_value:.2f}%")
```

```python
highest_decline = monthly_spending["Growth (%)"].idxmin()
decline_value = monthly_spending.loc[highest_decline, "Growth (%)"]

print("Highest Decline Category:", highest_decline)
print(f"Decline: {decline_value:.2f}%")
```

```python
monthly_spending.round(2)
```

# Feature 6 – Time-of-Day Analysis

```python
df["Hour"] = df["Time"].str[:2].astype(int)
```

```python
df[["Time", "Hour"]].head()
```

```python
# Hourly Spending

hourly_spending = (
    df[df["Type"] == "debit"]
    .groupby("Hour")["Amount"]
    .sum()
)

hourly_spending
```

```python
peak_hour = hourly_spending.idxmax()

print("Peak Spending Hour:", peak_hour)
```

```python
# Category-wise Hourly Spending

category_hour = (
    df[df["Type"] == "debit"]
    .pivot_table(
        values="Amount",
        index="Category",
        columns="Hour",
        aggfunc="sum",
        fill_value=0
    )
)

category_hour
```

```python
food_delivery = df[
    (df["Category"] == "Food Delivery") &
    (df["Type"] == "debit")
]
```

```python
late_night = food_delivery[
    (food_delivery["Hour"] >= 21) |
    (food_delivery["Hour"] <= 1)
]
```

```python
late_night_percentage = (
    len(late_night) / len(food_delivery)
) * 100

print(f"Late Night Food Delivery Orders: {late_night_percentage:.2f}%")
```

```python
peak_hour_category = (
    df[df["Type"] == "debit"]
    .groupby("Category")["Hour"]
    .agg(lambda x: x.mode()[0])
)

peak_hour_category
```

```python
print("=" * 50)
print("TIME OF DAY ANALYSIS")
print("=" * 50)

print(f"Peak Spending Hour: {peak_hour}:00")
print(f"Late Night Food Delivery Orders: {late_night_percentage:.2f}%")
```

# Feature 7 – Anomaly Detection

```python
df["Category_Mean"] = (
    df.groupby("Category")["Amount"]
      .transform("mean")
)
```

```python
df["Category_STD"] = (
    df.groupby("Category")["Amount"]
      .transform("std")
)
```

```python
df["Z_Score"] = (
    (df["Amount"] - df["Category_Mean"])
    / df["Category_STD"]
)
```

```python
anomalies = df[df["Z_Score"] > 2]
```

```python
anomalies = anomalies.sort_values(
    by="Z_Score",
    ascending=False
)
```

```python
anomalies[
    [
        "Date",
        "Vendor",
        "Category",
        "Amount",
        "Z_Score"
    ]
].head(10)
```

```python
print("Total Anomalies Detected:", len(anomalies))
```

```python
print("=" * 50)
print("ANOMALY DETECTION")
print("=" * 50)

print("Total Anomalies:", len(anomalies))

print("\nTop 5 Anomalies\n")

print(
    anomalies[
        [
            "Date",
            "Vendor",
            "Category",
            "Amount",
            "Z_Score"
        ]
    ].head()
)
```

# Feature 8 – Spending Archetype Detection

```python
archetypes = []

total_debit = df[df["Type"] == "debit"]["Amount"].sum()

```

```python
# FOODIE
food_spending = df[
    (df["Category"].isin(["Food Delivery", "Restaurants", "Cafe"])) &
    (df["Type"] == "debit")
]["Amount"].sum()

food_percentage = (food_spending / total_debit) * 100

if food_percentage > 25:
    archetypes.append("🍔 THE FOODIE")



# QUICK COMMERCE
quick_spending = df[
    (df["Category"] == "Quick Commerce") &
    (df["Type"] == "debit")
]["Amount"].sum()

quick_percentage = (quick_spending / total_debit) * 100

if quick_percentage > 15:
    archetypes.append("🛒 THE QUICK COMMERCE JUNKIE")



# SHOPAHOLIC
shopping = df[
    (df["Category"] == "E-Commerce") &
    (df["Type"] == "debit")
]["Amount"].sum()

shopping_percentage = (shopping / total_debit) * 100

if shopping_percentage > 15:
    archetypes.append("🛍️ THE SHOPAHOLIC")


# INVESTOR
investment = df[
    (df["Category"] == "Investments") &
    (df["Type"] == "debit")
]["Amount"].sum()

investment_percentage = (investment / total_debit) * 100

if investment_percentage > 15:
    archetypes.append("📈 THE INVESTOR")



# LATE NIGHT SNACKER
food_delivery = df[
    (df["Category"] == "Food Delivery") &
    (df["Type"] == "debit")
]

late_night = food_delivery[
    (food_delivery["Hour"] >= 21) |
    (food_delivery["Hour"] <= 1)
]

late_percentage = (len(late_night) / len(food_delivery)) * 100

if late_percentage > 50:
    archetypes.append("🌙 THE LATE-NIGHT SNACKER")



# SUBSCRIPTIONS
subscription_count = df[
    df["Category"] == "Subscriptions"
]["Vendor"].nunique()

if subscription_count >= 5:
    archetypes.append("🎵 THE SUBSCRIPTION LOVER")



# YOLO
if savings_rate < 10:
    archetypes.append("💸 THE YOLO SPENDER")



# DISCIPLINED
if savings_rate > 40:
    archetypes.append("💰 THE DISCIPLINED SAVER")
```

```python
print("=" * 50)
print("SPENDING ARCHETYPES")
print("=" * 50)

for archetype in archetypes:
    print(archetype)
```
